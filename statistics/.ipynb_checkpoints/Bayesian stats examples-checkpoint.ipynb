```python
# Example 1: Flipping a coin
# Updating beliefs about the probability for landing heads (H) based on collecting data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, binom

# We model our prior as a beta distribution (why?)

# Prior parameters
alpha_prior = 1
beta_prior = 1

# θ values between 0 and 1
theta = np.linspace(0, 1, 200)

# Prior PDF
prior_pdf = beta.pdf(theta, alpha_prior, beta_prior)

plt.plot(theta, prior_pdf, label='Prior Beta(1,1)')
plt.title('Prior Distribution for θ')
plt.xlabel('θ (probability of heads)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Let's say we flip the coin 10 times and get 7 heads

n = 10
x = 7  # number of heads

# Likelihood (up to proportionality)
likelihood = binom.pmf(x, n, theta)

plt.plot(theta, likelihood, label=f'Likelihood (x={x}, n={n})')
plt.title('Likelihood of θ given data')
plt.xlabel('θ')
plt.ylabel('Likelihood')
plt.legend()
plt.show()
```

```python
# Update our belief to posterior based on the observed likelihood

alpha_post = alpha_prior + x
beta_post = beta_prior + n - x

posterior_pdf = beta.pdf(theta, alpha_post, beta_post)

plt.plot(theta, prior_pdf, '--', label='Prior Beta(1,1)')
plt.plot(theta, posterior_pdf, label=f'Posterior Beta({alpha_post},{beta_post})')
plt.title('Prior vs Posterior for θ')
plt.xlabel('θ')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Let's run one more experiment

# Posterior from first experiment (this becomes the new prior)
alpha_prior_2 = alpha_post
beta_prior_2 = beta_post

# Flip a coin 14 times and get 5 heads
n = 14
x = 5  # number of heads

# Likelihood (up to proportionality)
likelihood = binom.pmf(x, n, theta)

plt.plot(theta, likelihood, label=f'Likelihood (x={x}, n={n})')
plt.title('Likelihood of θ given data')
plt.xlabel('θ')
plt.ylabel('Likelihood')
plt.legend()
plt.show()
```

```python
alpha_post_2 = alpha_prior_2 + x
beta_post_2 = beta_prior_2 + n - x

# Compute prior and posterior PDFs
prior_pdf = beta.pdf(theta, alpha_prior_2, beta_prior_2)
posterior_pdf = beta.pdf(theta, alpha_post_2, beta_post_2)

plt.plot(theta, prior_pdf, '--', label=f'Prior Beta({alpha_prior_2},{beta_prior_2})')
plt.plot(theta, posterior_pdf, label=f'Posterior Beta({alpha_post_2},{beta_post_2})')
plt.title('Prior vs Posterior after Second Experiment')
plt.xlabel('θ (probability of heads)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Sequential Bayesian Updating Simulation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, binom

# --- Simulation parameters ---
np.random.seed(42)
n = 20                    # number of flips per experiment
num_experiments = 8     # number of sequential experiments
heads_range = (7, 13)     # roughly near 50% heads

# --- Initial prior ---
alpha, beta_param = 1, 1
theta = np.linspace(0, 1, 300)
colors = plt.cm.coolwarm(np.linspace(0, 1, num_experiments + 1))

# --- For storing metrics ---
posterior_means = []
ci_lower, ci_upper = [], []
x_obs = []

# --- Plot posterior distributions ---
plt.figure(figsize=(8, 5))
prior_pdf = beta.pdf(theta, alpha, beta_param)
plt.plot(theta, prior_pdf, '--', color=colors[0], label=f'Prior Beta({alpha},{beta_param})')

for i in range(1, num_experiments + 1):
    # Generate data: random number of heads near 50%
    x = np.random.randint(*heads_range)
    x_obs.append(x)

    # Update posterior
    alpha += x
    beta_param += n - x

    # Posterior PDF
    posterior_pdf = beta.pdf(theta, alpha, beta_param)
    plt.plot(theta, posterior_pdf, color=colors[i], label=f'Exp {i}: {x}/{n} heads')

    # Track summary statistics
    mean = alpha / (alpha + beta_param)
    ci = beta.ppf([0.025, 0.975], alpha, beta_param)
    posterior_means.append(mean)
    ci_lower.append(ci[0])
    ci_upper.append(ci[1])

# --- Posterior curves ---
plt.title('Sequential Bayesian Updates (Blue → Red)')
plt.xlabel('θ (probability of heads)')
plt.ylabel('Density')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.grid(alpha=0.3)
plt.show()

# --- Plot posterior mean and credible interval evolution ---
experiments = np.arange(1, num_experiments + 1)

plt.figure(figsize=(8, 5))
plt.plot(experiments, posterior_means, '-o', label='Posterior mean', color='black')
plt.fill_between(experiments, ci_lower, ci_upper, color='gray', alpha=0.3, label='95% credible interval')
plt.title('Evolution of Posterior Mean and 95% Credible Interval')
plt.xlabel('Experiment number')
plt.ylabel('θ estimate')
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# --- Print final summary ---
print(f"Final Posterior: Beta({alpha}, {beta_param})")
print(f"Posterior mean = {posterior_means[-1]:.3f}")
print(f"95% credible interval = ({ci_lower[-1]:.3f}, {ci_upper[-1]:.3f})")

```

```python
# Example 2: Rolling a die
# Updating the beliefs about the probabilities of landing each side given collected data

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import dirichlet

# Sides of the die
sides = np.array([1, 2, 3, 4, 5, 6])

# Prior belief (theta)
theta_prior = np.array([1/9, 2/9, 1/9, 2/9, 2/9, 1/9])

# Plotting the prior
plt.bar(sides, theta_prior, color='skyblue')
plt.xlabel('Die Side')
plt.ylabel('Prior Probability')
plt.title('Prior Distribution over Die Sides')
plt.ylim(0, max(theta_prior)*1.2)
plt.show()
```

```python
# Observed counts from 20 rolls
observed_counts = np.array([2, 4, 1, 5, 6, 2])
total_rolls = observed_counts.sum()

# Calculate likelihood (normalized counts)
likelihood = observed_counts / total_rolls

# Plot likelihood
plt.bar(sides, likelihood, color='orange')
plt.xlabel('Die Side')
plt.ylabel('Likelihood (Observed Frequency)')
plt.title('Likelihood based on Experiment')
plt.ylim(0, max(likelihood)*1.2)
plt.show()
```

```python
# Update the posterior belief

# Prior strength (pseudo counts)
prior_strength = 9
alpha_prior = theta_prior * prior_strength

# Posterior alpha parameters
alpha_posterior = alpha_prior + observed_counts

# Calculate posterior mean (expected probabilities)
theta_posterior = alpha_posterior / alpha_posterior.sum()

# Plot prior and posterior side-by-side
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].bar(sides, theta_prior, color='skyblue')
ax[0].set_title('Prior Distribution')
ax[0].set_xlabel('Die Side')
ax[0].set_ylabel('Probability')
ax[0].set_ylim(0, max(theta_posterior)*1.2)

ax[1].bar(sides, theta_posterior, color='green')
ax[1].set_title('Posterior Distribution')
ax[1].set_xlabel('Die Side')
ax[1].set_ylabel('Probability')
ax[1].set_ylim(0, max(theta_posterior)*1.2)

plt.show()
```

```python
# Sequential Bayesian Updating Simulation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters (same as before)
sides = np.array([1, 2, 3, 4, 5, 6])
prior_theta = np.array([1/9, 2/9, 1/9, 2/9, 2/9, 1/9])
prior_strength = 9
alpha_prior = prior_theta * prior_strength
true_theta = np.ones(6) / 6
n_rolls = 20
n_experiments = 15

posteriors = [alpha_prior.copy()]

for i in range(n_experiments):
    rolls = np.random.choice(sides, size=n_rolls, p=true_theta)
    counts = np.array([np.sum(rolls == side) for side in sides])
    alpha_posterior = posteriors[-1] + counts
    posteriors.append(alpha_posterior)

posterior_means = np.array([alpha / alpha.sum() for alpha in posteriors])

# Compute 95% credible intervals for each side at each update
# Marginal for each side is Beta(alpha_i, sum(alpha) - alpha_i)
lower_bounds = []
upper_bounds = []

for alpha in posteriors:
    total_alpha = alpha.sum()
    lower = []
    upper = []
    for a_i in alpha:
        l = beta.ppf(0.025, a_i, total_alpha - a_i)
        u = beta.ppf(0.975, a_i, total_alpha - a_i)
        lower.append(l)
        upper.append(u)
    lower_bounds.append(lower)
    upper_bounds.append(upper)

lower_bounds = np.array(lower_bounds)
upper_bounds = np.array(upper_bounds)

# Plot posterior means with credible intervals over time for each side
plt.figure(figsize=(14, 9))

for idx, side in enumerate(sides):
    plt.plot(posterior_means[:, idx], marker='o', label=f'Side {side}')
    plt.fill_between(range(len(posteriors)), 
                     lower_bounds[:, idx], 
                     upper_bounds[:, idx], 
                     alpha=0.2)

plt.axhline(1/6, color='black', linestyle='dashed', label='True probability (1/6)')
plt.xlabel('Experiment Number')
plt.ylabel('Probability')
plt.title('Sequential Bayesian Updates with 95% Credible Intervals')
plt.legend()
plt.show()

```

```python
# Example 3: A normally distributed random variable X


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skewnorm

# Step 1: Define true distribution parameters
true_mu = 5.0
true_sigma = 1.0

# Step 2: Define skewed prior for mu using skewnorm
# skewnorm(a, loc, scale) with a>0 is positively skewed
a = 5  # skewness parameter
prior_loc = 3
prior_scale = 1.0

# Define grid for mu
mu_grid = np.linspace(0, 10, 500)

# Prior density
prior_pdf = skewnorm.pdf(mu_grid, a, loc=prior_loc, scale=prior_scale)

# Step 3: Generate experiment data
n_samples = 30
data = np.random.normal(true_mu, true_sigma, size=n_samples)

# Compute likelihood for each mu in grid:
# Since samples are independent,
# likelihood = product of N(data_i | mu, sigma^2)
# Work in log space for numerical stability
log_likelihood = np.sum(norm.logpdf(data[:, None], loc=mu_grid, scale=true_sigma), axis=0)
likelihood = np.exp(log_likelihood - np.max(log_likelihood))  # normalize to avoid underflow

# Step 4: Compute unnormalized posterior
unnormalized_posterior = prior_pdf * likelihood

# Normalize posterior
posterior_pdf = unnormalized_posterior / np.trapezoid(unnormalized_posterior, mu_grid)

# Plot prior, likelihood, posterior
plt.figure(figsize=(12, 7))

plt.plot(mu_grid, prior_pdf / prior_pdf.max(), label='Prior (scaled)', color='blue')
plt.plot(mu_grid, likelihood / likelihood.max(), label='Likelihood (scaled)', color='orange')
plt.plot(mu_grid, posterior_pdf / posterior_pdf.max(), label='Posterior (scaled)', color='green')

plt.axvline(true_mu, color='black', linestyle='--', label=r'True $\mu$')
plt.xlabel(r'$\mu$')
plt.ylabel('Density (scaled)')
plt.title('Bayesian Update with Skewed Prior and Normal Likelihood')
plt.legend()
plt.show()
```

```python
# Sequential Bayesian Updating Simulation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skewnorm
from matplotlib import colormaps

# True parameters
true_mu = 5.0
true_sigma = 1.0

# Prior parameters (skewed prior)
a = 5  # skewness
prior_loc = 3
prior_scale = 1.0

# Grid for mu
mu_grid = np.linspace(0, 10, 500)

# Initial prior pdf
prior_pdf = skewnorm.pdf(mu_grid, a, loc=prior_loc, scale=prior_scale)

# Number of sequential experiments
n_experiments = 1
n_samples_per_experiment = 5

# Store posterior pdfs
posterior_pdfs = [prior_pdf]

# Current prior to update in each iteration
current_prior = prior_pdf

for i in range(n_experiments):
    # Generate new data
    data = np.random.normal(true_mu, true_sigma, n_samples_per_experiment)
    
    # Compute log likelihood for each mu in grid
    log_likelihood = np.sum(norm.logpdf(data[:, None], loc=mu_grid, scale=true_sigma), axis=0)
    
    # Convert to likelihood, normalize for stability
    likelihood = np.exp(log_likelihood - np.max(log_likelihood))
    
    # Update unnormalized posterior
    unnorm_posterior = current_prior * likelihood
    
    # Normalize posterior
    posterior = unnorm_posterior / np.trapezoid(unnorm_posterior, mu_grid)
    
    # Store posterior
    posterior_pdfs.append(posterior)
    
    # Update current prior for next iteration
    current_prior = posterior

# Plot with gradient from blue to red
colors = colormaps['coolwarm'].resampled(n_experiments + 1) # +1 for prior

plt.figure(figsize=(12, 8))
for i, pdf in enumerate(posterior_pdfs):
    plt.plot(mu_grid, pdf / pdf.max(), color=colors(i), label=f'After experiment {i}')

plt.axvline(true_mu, color='black', linestyle='--', label=r'True $\mu$')
plt.xlabel(r'$\mu$')
plt.ylabel('Density (scaled)')
plt.title('Sequential Bayesian Updates: Prior to Final Posterior')
plt.legend(loc='upper left')
plt.show()
```
