```python
import numpy as np
from scipy.optimize import root_scalar

def constants_at_T(T_celsius=37.0):
    """Return temperature-dependent constants at T (°C)."""
    T_kelvin = T_celsius + 273.15

    # Kw (mol^2/L^2) – empirical fit
    logKw = -4470.99 / T_kelvin + 6.0875
    Kw = 10**logKw

    # CO2 solubility alpha (mol/(L·mmHg))
    alpha = (0.0307 - 0.00057*(T_celsius-37)) / 1000

    # K1' from pKa
    pK1 = 6.58 - 0.010*(T_celsius-37)
    K1prime = 10**(-pK1)

    # Weak acid Ka (approx; keep ~6.8)
    pKa = 6.8
    Ka = 10**(-pKa)

    return Kw, alpha, K1prime, Ka

def stewart_pH(
    PCO2=40.0, SID=42.0, Atot=15.0,
    T_celsius=37.0
):
    """Compute pH including T-dependence."""

    # Convert mmol/L -> mol/L
    SID /= 1000
    Atot /= 1000

    # Get constants at temperature
    Kw, alpha, K1prime, Ka = constants_at_T(T_celsius)

    # f(H) = electroneutrality
    def f(H):
        HCO3 = (K1prime * alpha * PCO2) / H
        OH = Kw / H
        A_minus = (Atot * Ka) / (Ka + H)
        return SID - (HCO3 + A_minus + OH - H)

    # Solve for H
    sol = root_scalar(f, x0=4e-8, method='newton')  # start near pH ~7.4
    if not sol.converged:
        raise RuntimeError("Root finding did not converge")
    H = sol.root
    pH = -np.log10(H)

    # Species
    HCO3 = (K1prime * alpha * PCO2) / H
    OH = Kw / H
    A_minus = (Atot * Ka) / (Ka + H)

    # Convert back to mmol/L
    HCO3 *= 1000
    OH *= 1000
    A_minus *= 1000
    H_conc = H * 1000

    return {
        "pH": pH,
        "T (°C)": T_celsius,
        "[H+] (mmol/L)": H_conc,
        "[HCO3-] (mmol/L)": HCO3,
        "[OH-] (mmol/L)": OH,
        "[A-] (mmol/L)": A_minus
    }

# Example: compare across different temperatures
for T in [30, 37, 42]:
    print(f"\n--- Results at {T} °C ---")
    result = stewart_pH(T_celsius=T)
    for k, v in result.items():
        if isinstance(v, float):
            print(f"{k:20s}: {v:.4f}")
        else:
            print(f"{k:20s}: {v}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Reuse your stewart_pH and constants_at_T definitions from before

def simulate_over_time(times, PCO2_func, SID_func, Atot_func, T_func):
    """Simulate Stewart variables over time."""
    results = {
        "time": times,
        "pH": [],
        "HCO3": [],
        "A_minus": [],
        "OH": [],
        "Hplus": [],
    }
    for t in times:
        res = stewart_pH(
            PCO2=PCO2_func(t),
            SID=SID_func(t),
            Atot=Atot_func(t),
            T_celsius=T_func(t)
        )
        results["pH"].append(res["pH"])
        results["HCO3"].append(res["[HCO3-] (mmol/L)"])
        results["A_minus"].append(res["[A-] (mmol/L)"])
        results["OH"].append(res["[OH-] (mmol/L)"])
        results["Hplus"].append(res["[H+] (mmol/L)"])
    return results

# Example scenario: 0–60 min
times = np.linspace(0, 600, 100)
PCO2_func = lambda t: 40 + (20 * t/60)   # rise 40→60 mmHg
SID_func  = lambda t: 42 - (4 * t/60)    # fall 42→38 mmol/L
Atot_func = lambda t: 15                 # constant mmol/L
T_func    = lambda t: 37                 # constant °C

results = simulate_over_time(times, PCO2_func, SID_func, Atot_func, T_func)

# Convert results to arrays for easier plotting
pH = np.array(results["pH"])
HCO3 = np.array(results["HCO3"])
A_minus = np.array(results["A_minus"])
OH = np.array(results["OH"])
Hplus = np.array(results["Hplus"])

# Plot in subplots
fig, axes = plt.subplots(5, 1, figsize=(10, 14), sharex=True)

axes[0].plot(times, pH, color='blue')
axes[0].set_ylabel("pH")
axes[0].set_title("Stewart Model Dynamics Over Time")
axes[0].grid(True)

axes[1].plot(times, HCO3, color='green')
axes[1].set_ylabel("[HCO3-] (mmol/L)")
axes[1].grid(True)

axes[2].plot(times, A_minus, color='orange')
axes[2].set_ylabel("[A-] (mmol/L)")
axes[2].grid(True)

axes[3].plot(times, OH, color='purple')
axes[3].set_ylabel("[OH-] (mmol/L)")
axes[3].grid(True)

axes[4].plot(times, Hplus, color='red')
axes[4].set_ylabel("[H+] (mmol/L)")
axes[4].set_xlabel("Time (minutes)")
axes[4].grid(True)

plt.tight_layout()
plt.show()
```
