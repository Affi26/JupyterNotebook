```python
import numpy as np
import matplotlib.pyplot as plt

# Engine geometry
V_c = 5e-5       # m^3, clearance volume
D = 0.086        # m, bore
R = 0.043        # m, crank radius
L = 0.128        # m, connecting rod length
gamma = 1.35     # specific heat ratio
rpm = 1000
n_crank = 720    # degrees per cycle
theta = np.linspace(0, 2*np.pi, n_crank)
dt = (60/(rpm*360)) * 2*np.pi  # s per radian

# Cylinder volume and derivative
V = V_c + (np.pi*D**2/4) * (R + L - R*np.cos(theta) - np.sqrt(L**2 - (R*np.sin(theta))**2))
dV = np.gradient(V, theta[1]-theta[0])

# True heat release (simplified)
Q_dot_true = 5e5 * np.exp(-((theta-np.pi)/0.2)**2)  # peak at TDC
p0 = 1e5
p_true = np.zeros(n_crank)
p_true[0] = p0

for k in range(1, n_crank):
    dp = (gamma-1)/V[k] * (Q_dot_true[k] - gamma*p_true[k-1]*dV[k]/(gamma-1))
    p_true[k] = p_true[k-1] + dp*dt

# Measurement noise
noise_std = 500
p_meas = p_true + np.random.normal(0, noise_std, n_crank)

# EKF setup
x_est = np.array([1e5, 0.0])
P = np.eye(2)*1e5
Q_proc = np.diag([1e2, 1e4])
R_meas = noise_std**2
x_hist = np.zeros((n_crank,2))

def f(x, V, dV, gamma, dt):
    p, Q_dot = x
    dpdt = (gamma-1)/V * (Q_dot - gamma*p*dV/(gamma-1))
    return np.array([p + dpdt*dt, Q_dot])

def jacobian_F(x, V, dV, gamma, dt):
    p, Q_dot = x
    dp_dp = 1 - (gamma*dV/V)*dt
    dp_dQ = (gamma-1)/V*dt
    F = np.array([[dp_dp, dp_dQ],
                  [0,1]])
    return F

# Run EKF
for k in range(n_crank):
    F = jacobian_F(x_est, V[k], dV[k], gamma, dt)
    x_pred = f(x_est, V[k], dV[k], gamma, dt)
    P_pred = F @ P @ F.T + Q_proc

    H = np.array([[1,0]])
    y = p_meas[k] - H @ x_pred
    S = H @ P_pred @ H.T + R_meas
    K = P_pred @ H.T / S

    x_est = x_pred + K.flatten() * y
    P = (np.eye(2) - K @ H) @ P_pred
    x_hist[k] = x_est

# Plot results
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(theta*180/np.pi, p_true, label='True Pressure')
plt.plot(theta*180/np.pi, p_meas, label='Measured Pressure', alpha=0.5)
plt.plot(theta*180/np.pi, x_hist[:,0], label='EKF Pressure')
plt.ylabel('Pressure [Pa]')
plt.legend()

plt.subplot(2,1,2)
plt.plot(theta*180/np.pi, Q_dot_true, label='True Heat Release')
plt.plot(theta*180/np.pi, x_hist[:,1], label='EKF Estimate')
plt.xlabel('Crank Angle [deg]')
plt.ylabel('Heat Release Rate [J/s]')
plt.legend()
plt.show()
```
